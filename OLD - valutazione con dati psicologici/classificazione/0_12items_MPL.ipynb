```python

def print_acc_loss(model,X_train,y_train,X_test,y_test):
    # Evaluate the model on train and test sets
    train_loss=model.evaluate(X_train, y_train, verbose=0)[0]
    test_loss=model.evaluate(X_test, y_test, verbose=0)[0]
    
    # Evaluate the model on train and test sets
    train_accuracy = model.evaluate(X_train, y_train, verbose=0)[1]
    test_accuracy = model.evaluate(X_test, y_test, verbose=0)[1]
    
    print(f"Final accuracy on train set: {train_accuracy:.4f}")
    print(f"Final accuracy on test set: {test_accuracy:.4f}")
    
    print(f"Final loss on train set: {train_loss:.4f}")
    print(f"Final loss on test set: {test_loss:.4f}")
    
   
def plot_confusion_matrix(model,X_train,y_train,X_test,y_test):    
   # Calculate and print confusion matrices
    y_train_pred = np.argmax(model.predict(X_train), axis=1)
    y_test_pred = np.argmax(model.predict(X_test), axis=1)
    
    train_confusion = confusion_matrix(y_train, y_train_pred)
    test_confusion = confusion_matrix(y_test, y_test_pred)
    
    # Create a 1x2 grid for subplots
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    
    # Plot confusion matrix for train set
    sns.heatmap(train_confusion, annot=True, linewidths=0.01, cmap="Greens", linecolor="gray", fmt='.1f', ax=axs[0])
    axs[0].set_xlabel("Predicted Label")
    axs[0].set_ylabel("True Label")
    axs[0].set_title("Confusion Matrix (Train Set)")
    
    # Plot confusion matrix for test set
    sns.heatmap(test_confusion, annot=True, linewidths=0.01, cmap="Greens", linecolor="gray", fmt='.1f', ax=axs[1])
    axs[1].set_xlabel("Predicted Label")
    axs[1].set_ylabel("True Label")
    axs[1].set_title("Confusion Matrix (Test Set)")
    
    plt.show()
    
def plot_performance(history):
       
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Test Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Test Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()    
```

```python

# Modello 1: 2 layers Dense Con 16 neuroni e 4 neuroni
def train_model_1(X, y, test_size, random_state, batch_value, num_epochs,lr):

    optimizer= Adam(learning_rate=lr)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    ann1 = Sequential([   
        Dense(16, input_shape=(12,), activation='relu'),  
        Dense(4, activation='softmax')  
    ])
    
    ann1.compile(optimizer=optimizer,  
                  loss='sparse_categorical_crossentropy',  
                  metrics=['accuracy'])  
    
    history = ann1.fit(X_train,y_train,
                      batch_size=batch_value,
                      epochs=num_epochs,
                      validation_data=(X_test, y_test),
                      verbose=2)   
    
    # funzione per stamapre acc e loss
    print_acc_loss(ann1,X_train,y_train,X_test,y_test)
    # funzione per plottare il confusion matrix
    plot_confusion_matrix(ann1,X_train,y_train,X_test,y_test)
    
    return history

# Modello 2: 2 layers Dense con 32 neuroni nel primo layer
def train_model_2(X, y, test_size, random_state, batch_value, num_epochs,lr):

    optimizer= Adam(learning_rate=lr)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    ann1 = Sequential([   
        Dense(32, input_shape=(12,), activation='relu'),  
        Dense(4, activation='softmax')  
    ])
    
    ann1.compile(optimizer=optimizer,  
                  loss='sparse_categorical_crossentropy',  
                  metrics=['accuracy'])  
    
    history = ann1.fit(X_train,y_train,
                      batch_size=batch_value,
                      epochs=num_epochs,
                      validation_data=(X_test, y_test),
                      verbose=2)   
    
    # funzione per stamapre acc e loss
    print_acc_loss(ann1,X_train,y_train,X_test,y_test)
    # funzione per plottare il confusion matrix
    plot_confusion_matrix(ann1,X_train,y_train,X_test,y_test)
    
    return history


# Modello 3: 3 layers Dense
def train_model_3(X, y, test_size, random_state, batch_value, num_epochs,lr):

    optimizer= Adam(learning_rate=lr)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    ann2 = Sequential([   
        Dense(16, input_shape=(12,), activation='relu'),  
        Dense(16, activation='relu'),
        Dense(4, activation='softmax')  
    ])
    
    ann2.compile(optimizer=optimizer,  
                  loss='sparse_categorical_crossentropy',  
                  metrics=['accuracy'])  
    
    history = ann2.fit(X_train,y_train,
                      batch_size=batch_value,
                      epochs=num_epochs,
                      validation_data=(X_test, y_test),
                      verbose=2)   
    
    # funzione per stamapre acc e loss
    print_acc_loss(ann2,X_train,y_train,X_test,y_test)
    # funzione per plottare il confusion matrix
    plot_confusion_matrix(ann2,X_train,y_train,X_test,y_test)
    
    return history

# Modello 4: 4 layers Dense
def train_model_4(X, y, test_size, random_state, batch_value, num_epochs,lr):

    optimizer= Adam(learning_rate=lr)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    ann3 = Sequential([   
        Dense(32, input_shape=(12,), activation='relu'),  
        Dense(16, activation='relu'),
        Dense(16, activation='relu'),
        Dense(4, activation='softmax')  
    ])
    
    ann3.compile(optimizer=optimizer,  
                  loss='sparse_categorical_crossentropy',  
                  metrics=['accuracy'])  
    
    history = ann3.fit(X_train,y_train,
                      batch_size=batch_value,
                      epochs=num_epochs,
                      validation_data=(X_test, y_test),
                      verbose=2)   
    
    # funzione per stamapre acc e loss
    print_acc_loss(ann3,X_train,y_train,X_test,y_test)
    # funzione per plottare il confusion matrix
    plot_confusion_matrix(ann3,X_train,y_train,X_test,y_test)
    
    return history
```

> Overfitting
- Segnali: Se la Training Loss continua a diminuire mentre la Validation Loss inizia ad aumentare dopo un certo punto, il modello è in overfitting.
- Accuratezza: La Training Accuracy sarà alta mentre la Validation Accuracy sarà significativamente più bassa.
> Underfitting
- Segnali: Se sia la Training Loss che la Validation Loss sono alte e non diminuiscono significativamente, il modello è in underfitting.
- Accuratezza: La Training Accuracy e la Validation Accuracy saranno entrambe basse.
